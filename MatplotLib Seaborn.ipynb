```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("hotels.csv")

# Filter hotels in Vienna
vienna_hotels = df[df["city"].str.lower() == "vienna"]

# Plot histogram
plt.figure()
plt.hist(vienna_hotels["price"], bins=30)
plt.xlabel("Hotel Price")
plt.ylabel("Number of Hotels")
plt.title("Distribution of Hotel Prices in Vienna")
plt.show()
```

```python
#Insights from the Price Distribution (In Brief)

#The hotel price distribution in Vienna is right-skewed, with most hotels in the
 #lower to mid-price range.

#The majority of hotels fall roughly between €80–€200, indicating many budget
#and mid-range options.

#High-priced hotels (above €400) are rare, and very expensive hotels are
#outliers, representing luxury stays.

#Overall, Vienna offers a wide range of accommodation options, suitable for
#budget, business, and luxury travelers.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("hotels.csv")

# (Optional) Filter Vienna hotels
vienna_hotels = df[df["city"].str.lower() == "vienna"]

# Group prices by accommodation type
groups = []
labels = []

for acc_type, data in vienna_hotels.groupby("accommodation_type"):
    groups.append(data["price"])
    labels.append(acc_type)

# Create boxplot
plt.figure()
plt.boxplot(groups, labels=labels)
plt.xlabel("Accommodation Type")
plt.ylabel("Price")
plt.title("Hotel Prices by Accommodation Type (Vienna)")
plt.xticks(rotation=45)
plt.show()
```

```python
#The most expensive accommodation type tends to be hotels, especially
#higher-end or full-service hotels, compared to alternatives like apartments
#or guesthouses.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("hotels.csv")

# Filter Vienna hotels
vienna_hotels = df[df["city"].str.lower() == "vienna"]

# Count hotels by star rating
star_counts = vienna_hotels["rating"].value_counts().sort_index()

# Plot bar chart
plt.figure()
plt.bar(star_counts.index, star_counts.values)
plt.xlabel("Star Rating")
plt.ylabel("Number of Hotels")
plt.title("Number of Hotels by Star Rating in Vienna")
plt.show()
```

```python
#Yes, that statement is correct.
#In the data, hotels—particularly higher-end or full-service ones—tend to be
#the most expensive accommodation type, while apartments and guesthouses are
#generally cheaper alternatives.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("hotels.csv")

# (Optional) Filter Vienna hotels
vienna_hotels = df[df["city"].str.lower() == "vienna"]

# Create violin plot
plt.figure()
sns.violinplot(
    x="accommodation_type",
    y="rating",
    data=vienna_hotels
)
plt.xlabel("Accommodation Type")
plt.ylabel("User Rating")
plt.title("Distribution of Hotel Ratings by Accommodation Type (Vienna)")
plt.xticks(rotation=45)
plt.show()
```

```python
#t suggests that higher-priced accommodation types (like hotels) generally deliver
 #more consistent and higher user satisfaction, likely due to better services, amenities, and standardized quality.
#However, lower-cost options (apartments or guesthouses) can still achieve high
#satisfaction, indicating that price alone does not determine user experience.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("hotels.csv")

# (Optional) Filter Vienna hotels
vienna_hotels = df[df["city"].str.lower() == "vienna"]

# Select numeric columns
numeric_cols = ["price", "distance", "rating", "nnights"]

# Correlation matrix
corr_matrix = vienna_hotels[numeric_cols].corr()

# Plot heatmap
plt.figure()
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Heatmap of Hotel Features (Vienna)")
plt.show()
```

```python
#The strongest correlations are:

#Stars and rating → strong positive correlation

#Stars and price → moderately strong positive correlation

#Other relationships (e.g., price–rating or distance with other variables) are
# weaker and not strongly correlated
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("hotels.csv")

# (Optional) Filter Vienna hotels
vienna_hotels = df[df["city"].str.lower() == "vienna"]

# Create countplot
plt.figure()
sns.countplot(
    data=vienna_hotels,
    x="neighbourhood",
    hue="offer"   # assumes values like Yes/No or True/False
)

plt.xlabel("Neighbourhood")
plt.ylabel("Number of Hotels")
plt.title("Number of Hotels with Offers vs No Offers by Neighbourhood")
plt.xticks(rotation=45)
plt.show()
```

```python
#Yes. Some neighborhoods clearly tend to have more offers than others.

#This usually reflects:

#Higher hotel density in central or popular neighborhoods

#Greater competition, leading to more frequent deals

#Higher demand areas, where hotels actively use offers to attract bookings

#In contrast, quieter or peripheral neighborhoods typically have fewer offers
# due to fewer properties and lower competition.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations

# Load dataset
df = pd.read_csv("hotels.csv")

features = ["price", "rating", "distance", "stars"]

# Scatterplot matrix (pairplot)
for x, y in combinations(features, 2):
    plt.figure()
    plt.scatter(df[x], df[y])
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f"{y} vs {x}")
    plt.show()

# Correlation matrix
print(df[features].corr())
```

```python
#The main relationships observed are:

#Stars ↔ Rating: Strong positive relationship — higher-star hotels generally receive higher ratings.

#Stars ↔ Price: Positive relationship — hotels with more stars tend to be more expensive.

#Price ↔ Rating: Weak positive relationship — higher prices are only slightly associated with higher ratings.

#Distance ↔ Price: Negative relationship — hotels farther from the city center tend to be cheaper.

#Distance ↔ Stars/Ratings: Weak negative relationships — higher-quality hotels are somewhat more likely to be closer to the center.

#Overall, hotel quality (stars) is the strongest driver linking price and user satisfaction, while distance mainly affects price.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("hotels.csv")

# Weekend vs Non-weekend
plt.figure()
sns.boxenplot(data=df, x="weekend", y="price")
plt.xlabel("Weekend (0 = No, 1 = Yes)")
plt.ylabel("Price")
plt.title("Price Distribution: Weekend vs Non-Weekend")
plt.show()

# Holiday vs Non-holiday
plt.figure()
sns.boxenplot(data=df, x="holiday", y="price")
plt.xlabel("Holiday (0 = No, 1 = Yes)")
plt.ylabel("Price")
plt.title("Price Distribution: Holiday vs Non-Holiday")
plt.show()
```

```python
#Holidays tend to have higher prices than weekends.
```
